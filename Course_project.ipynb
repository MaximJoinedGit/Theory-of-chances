## <center>Тема: проверка гипотезы о глобальном потеплении</center>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

#### Исходный архив содержит данные о температурных наблюдениях в 3448 городах мира. Скачан был с [Kaggle](https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data).

#### В архиве несколько файлов. Возьмем файл с данными только по крупным городам.

```python
path = 'GlobalLandTemperaturesByMajorCity.csv'
df = pd.read_csv(path)
```

#### Выясним, какие из них имеют самые полные данные, где меньше всего пропусков

```python
full_data_per_city = df.groupby(['dt', 'City']).mean()['AverageTemperature'].unstack(level=1).isnull().sum().sort_values()/df['dt'].nunique()
```

#### Построим графики изменения температур

```python
temperature_by_city = df.groupby(['dt', 'City']).mean()['AverageTemperature'].unstack(level=1)
```

```python
twelve_cities = full_data_per_city[:12].index.tolist()

temperature_by_city[twelve_cities].plot(figsize=(20,40), subplots=True)
plt.show()
```

#### Из графиков такого масштаба невозможно сделать выводы. Попробуем сгруппировать значения по десятилетиям (среднее значение за 10 лет) и построить графики по этим данным. В данных до 1750 года есть пробелы, поэтому начнем ровно с 1750 года.

```python
df['dt'] = [datetime.strptime(date, '%Y-%m-%d') for date in df['dt']]
df['decade'] = df['dt'].apply(lambda x: x.year//10)
dec_temperature_by_city = df.groupby(['decade', 'City']).mean()['AverageTemperature'].unstack(level=1)[1:]
```

```python
dec_temperature_by_city[twelve_cities].plot(figsize=(20,40), subplots=True, grid=True, marker='8',)
plt.show()
```

```python
data_to_manual_calc = dict()

for city in twelve_cities:
    data_to_manual_calc[city] = list(round(df.groupby(['City', 'decade']).mean()['AverageTemperature'].loc[city], 3))
```

#### Напишем функцию для проверки гипотезы о глобальном потеплении:

#### Формула для нахождения доверительного интервала для мат. ожидания будет иметь вид: $\overline{x} - t_{\gamma} \cdot \frac{S}{\sqrt{n}} < M(x) < \overline{x} + t_{\gamma} \cdot \frac{S}{\sqrt{n}}$

#### По результатам нам нужно выяснить, входит ли значение средней температуры для последнего десятилетия в доверительный интервал для матожидания с надежностью 0.95

#### По таблице критических значений t-критерия Стьюдента найдем $t_{\gamma} = 2,0518$ для n = 27 и p = 0,95

```python
def check_hypothesis(measurments: list):
    n = 27
    meas_mid = np.sum(measurments[:-1]) / n
    s = np.sqrt(np.sum([(num - meas_mid)**2 for num in measurments[:-1]]) / n)
    t_1 = meas_mid - 2.0518 * s / np.sqrt(n)
    t_2 = meas_mid + 2.0518 * s / np.sqrt(n)
    return t_1, t_2
```

```python
for city, decade_temperatures in data_to_manual_calc.items():
    check_hypothesis(decade_temperatures)
    print(f'{city}.\nДоверительные интервалы для матожидания:\nот {check_hypothesis(data_to_manual_calc[city])[0]} до {check_hypothesis(data_to_manual_calc[city])[1]},\nзначение за последнее десятилетие: {data_to_manual_calc[city][-1]}\n')
    print(f'Значение за последнее выходит за рамки доверительного интервала? {data_to_manual_calc[city][-1] > check_hypothesis(data_to_manual_calc[city])[1]}\n')
```

### Таким образом для всех городов, где были собраны наиболее полные данные о температурах с 1750 года мы видим, что средняя температура за последнее десятилетие превышает доверительный интервал для матожидания. Следовательно, гипотеза о глобальном потеплении доказана.
