### 1. Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

#### $n = 256; \;\; \gamma = 0.95; \;\; S_0 = 16; \;\; \overline{x} = 80$

#### $T_{1,2} = \overline{x} \pm \frac{S_0}{\sqrt{n}} \cdot C_{\gamma} = 80 \pm \frac{16}{\sqrt{256}} \cdot 1,96$

#### $T_1 = 78,04; \;\; T_2 = 81,96$

### 2. В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1
### Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
x = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
n = len(x)
M_x = sum(x) / len(x)
S_0 = ((sum([(num - M_x)**2 for num in x]))/n)**0.5
print(f'Исходя из данных испытаний мы получили количество испытаний: {n}, выборочную среднюю: {M_x} и среднеквадратическое отклонение: {S_0}')
```

#### $n = 10; \;\; \overline{x} = 6,59, \;\; S_0 = 0.4276$

#### $T_{1,2} = \overline{x} \pm \frac{S_0}{\sqrt{n}} \cdot C_{\gamma} = 6,59 \pm \frac{0,4276}{\sqrt{10}} \cdot 1,96$

#### Получается, что истинное значение величины Х с доверительной вероятностью 0,95 лежит в интервале от $T_1 = 6,325$ до $T_2 = 6,855$

### 3. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.

#### $\sigma^2 = 4 mm; \;\; \alpha = 0,05; \;\; M_{doc}(x) = 17 mm; \;\; n = 100; \;\; M_{real}(x) = 17,5 mm$

#### $H_0: M(x) = 17$

#### $H_1: M(x) > 17$

#### $t = \frac{M_{real}(x) - M_{doc}(x)}{\sigma} = \frac{17,5 - 17}{2} = 0,25$

#### Табличное значение t-критерия с односторонней критической областью для n = 10 равно 1,812. Полученное значение значительно меньше критического, следовательно, нулевая гипотеза верна.

### 4. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

```python
x = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
n = 10
p = 0.99
M_x = sum(x) / n
S_0 = ((sum([(num - M_x)**2 for num in x]))/n)**0.5
print(f'Из данных мы получили: выборочную среднюю: {M_x} и среднеквадратическое отклонение: {S_0}')
```

#### $H_0: M(x) = 200$

#### $H_1: M(x) \ne 200$

#### $t = \frac{M_{real}(x) - M_{doc}(x)}{\sigma} = \frac{198,5 - 200}{4,2249} = -0,355$

#### Табличное значение t-критерия с двусторонней критической областью для n = 10 и доверительной вероятностью 0,99 равно $\pm3,169$. Полученное значение значительно меньше критического, следовательно, нулевая гипотеза верна.
